```python
#!__*__coding:utf-8__*__

import pandas as pd

df_star = pd.read_excel("스타벅스거르기2.xlsx")
df2 = df_star[["상호명", "시도명", "시군구명"]]
gf = df2.groupby(["상호명", "시도명"]).count()
gf = gf.reset_index()
gf.to_excel("스타벅스_상점수.xlsx")

for i in gf.columns:
    print(i)

print(gf.head())

print(gf.info())
```
